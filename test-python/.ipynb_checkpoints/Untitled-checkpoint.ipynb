```python

def clean_file(name_file_input):
    file_input = open(name_file_input, "r+")
    s = ""
    for i in file_input:
        s = s+i
    s = re.sub(r'\n[\s\n]+', "\n", s)
    s = re.sub(r'[\t ]{2,}', " ", s)
    file_input.seek(0)
    file_input.write(s)
    file_input.close()
```
